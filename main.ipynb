### Morocco population between 2000 and 2025

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 

# read the file
df = pd.read_csv("./data/morocco_population.csv")

sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="Year", y="People", marker="o", color="#EF7722")
# plt.fill_between(df["Year"], df["People"], alpha=0.1, color="#007ACC")
plt.title("Morocco population", pad=20, color='#0BA6DF', size="24")


plt.show()
```
